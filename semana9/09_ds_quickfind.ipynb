```python
import graphviz as gv
```

```python
def drawDS(ds):
  graph = gv.Digraph("DisjointSet") # -> tipo de grafo ya definido en la libreria
  graph.graph_attr['rankdir'] = "BT" # -> ordenamiento del grafo
  for e, p in enumerate(ds):
    graph.node(str(e))
    if p != e:
      graph.edge(str(e), str(p))
  return graph
```

```python
ds = [0, 1, 2, 2, 8, 5, 5, 7, 8, 2] # -> cuando un nodo es hijo, tiene el numero de su padre en lugar del propio numero.
# Ej: 3 es hijo de 2, ya que donde deberia de estar 3, hay un 2...
drawDS(ds)
```

# Quick Find

```python
def find(ds, e):
  return ds[e]
```

```python
find(ds, 0), find(ds, 7), find(ds, 3), find(ds, 9), find(ds, 2) # -> al hacer find se busca al nodo o a su padre (si tiene uno)
```

```python
def union(ds, a, b): # -> ds: grafo, a: nodo 1, b: nodo b
  # Primero los va a buscar
  a = find(ds,  a)
  b = find(ds, b)

  if a == b: return # -> si son iguales entonces no devuelve nada

  for e, p in enumerate(ds): # -> itera sobre el grafo con indices y valores. 
    if p == a: # -> si el valor actual es igual al que pasas como parametro
      ds[e] = b # -> entonces cambia el indice de ese nodo para que tenga el valor de b , generando asi que sea su hijo.
```

```python
ds = [i for i in range(10)]
drawDS(ds)
```

```python
union(ds, 3, 4)
drawDS(ds)
```
